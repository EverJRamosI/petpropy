### Importations

```python
import numpy as np
import matplotlib.pyplot as plt
from petpropy import sigma_w, Rsw, c_w, rho_w, mu_w, Bw
```

```python
pressures = np.arange(500, 10500, 500) #psia
temperature = 200 + 460 #oR
sal = 20000 #ppm
```

### Solution Gas-Water Ratio

```python
rsw = Rsw.culberson_macketta(pressures, temperature, S=sal)
print(rsw)
```

```python
plt.plot(pressures, rsw)
plt.title('Solution Gas-Water Ratio')
plt.xlabel('Pressures')
plt.ylabel('Rsw')
plt.show()
```

### Water Volumetric Factor

```python
bw = Bw.mccain(pressures, temperature)
print(bw)
```

```python
plt.plot(pressures, bw)
plt.title('Water Volumetric Factor')
plt.xlabel('Pressures')
plt.ylabel('Bw')
plt.show()
```

### Water Compressibility

```python
cw = c_w.osif(pressures, temperature, S=sal)
print(cw)
```

```python
plt.plot(pressures, cw)
plt.title('Water Compressibility')
plt.xlabel('Pressures')
plt.ylabel('cw')
plt.show()
```

### Water Viscosity

```python
muw = mu_w.mccain(pressures, temperature, S=sal)
print(muw)
```

```python
plt.plot(pressures, muw)
plt.title('Water Viscosity')
plt.xlabel('Pressures')
plt.ylabel('Viscosity')
plt.show()
```

### Water Density

```python
rhow = rho_w.mccain(bw, S=sal)
print(rhow)
```

```python
plt.plot(pressures, rhow)
plt.title('Water Density')
plt.xlabel('Pressures')
plt.ylabel('Density')
plt.show()
```
